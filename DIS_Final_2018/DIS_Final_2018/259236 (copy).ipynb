### Definition

In this problem, you will have to implement node betweeness centrality, a measure that provide a score for every node $n$ of a graph $G$.  
  
Betweenness centrality $c_B(n)$ of a node $n$ is the sum of the fraction of all-pairs shortest paths that pass through $n$:  
  
$\Large c_B(n) =\sum_{s,t \in N} \frac{\sigma(s, t|n)}{\sigma(s, t)}$
  
where $N$ is the set of nodes, $\sigma(s, t)$ is the number of shortest paths connecting $s$ and $t$, and $\sigma(s, t|n)$ is the number of paths passing through node $n$, provided $n$ is different from $s$ and $t$. For the other cases $\sigma(s, t|n)$ is defined as follows: 
* If $s = t$: $\sigma(s, t) = 1$
* If $n \in \{s, t\}$: $\sigma(s, t|n) = 0$

Note that this definition of node betweenness employs no normalization.

For PageRank, consider that every edge of the undirected graph is replaced by two directed edges going in both directions. Provide an answer and an argument to support your answer in the cell below. 

Answer:

No, the PageRank algorithm does not produce the same values as node betweenness. PageRank gives higher values to nodes with higher numbers of incoming and outgoing edges. On the other hand, betweenness is a measure for how well edges separate communities. However, in some cases (such as if the nodes are connected in a ring or a clique) the betweenness will be the same for all nodes in the graph and the PageRank values will also be the same for all nodes.

Whereas if we had something like 

`clique1 -- x -- y -- z -- clique2` (where x, y, and z are nodes), y would have a high node betweenness but would not be rated so high by PageRank.

```python
characters = list(G.nodes())
result = {}

# compute a score for every character
for n in characters:

    # accumulator for the score of the current character
    accum = 0

    # shortest path FROM every character...
    for i,s in enumerate(characters):
        # ..TO every character
        for j,t in enumerate(characters):
            # we do not consider the same path twice
            if j > i:
                continue
            
            try:
                # shortest paths from s to t              
                all_short = nx.all_shortest_paths(G,s,t)
                
                ### YOUR CODE HERE ###################                
                sigma_st = 1
                sigma_st_n = 0
                
                if n != s and n != t:
                    list_all_short = [x for x in all_short]
                    sigma_st = len(list_all_short)
                    list_n_short = [x for x in list_all_short if n in x]
                    sigma_st_n = len(list_n_short)
                    accum += 1.0*sigma_st_n/sigma_st
                
            except nx.NetworkXNoPath:
                continue
    result[n] = accum

ex1_print_topn(result,10)
```

```python
#Required libraries.
import os
import string
import numpy as np
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
def read_corpus():
    '''Reads corpus from files.'''
    
    neg_reviews = []
    pos_reviews = []
    DIR = 'movie_reviews/'

    for l, sent in [(neg_reviews, 'neg'), (pos_reviews, 'pos')]:
        for file in os.listdir(DIR+sent):
            with open(os.path.join(DIR+sent, file)) as f:
                text = f.read()
            # split into words
            tokens = word_tokenize(text)
            # convert to lower case
            tokens = [w.lower() for w in tokens]
            # remove punctuation from each word
            table = str.maketrans('', '', string.punctuation)
            stripped = [w.translate(table) for w in tokens]
            # remove remaining tokens that are not alphabetic
            words = [word for word in stripped if word.isalpha()]
            # filter out stop words
            words = [w for w in words if not w in stop_words]

            l.append(' '.join(words))
    return neg_reviews, pos_reviews
```

### Task 2
Transform the reviews of the training set into vectors using the Bag-Of-Words model. Each dimension of a vector must count the number of occurences of a word in the review.
* _Hint 1: Make sure that all the vectors belong to the same space._
* _Hint 2: For this task you can reuse code from the first exercise of DIS2018 (Vector Space Retrieval)._

```python
def create_vocabulary(corpus, min_tf):
    '''Creates the vocabulary (set of distinct words) that appear more than $min_tf$ times in the corpus.'''
    
    count = {}
    for document in corpus:
        for word in document.split():
            if word in count:
                count[word] += 1
            else:
                count[word] = 0
    
    vocabulary = []
    for w, c in count.items():
        if c > min_tf:
            vocabulary.append(w)
    
    return vocabulary
    
vocabulary = create_vocabulary(reviews_train, len(reviews_train)*0.005)
```

```python
def transform_to_vector(reviews):
    '''Transforms a list of reviews to a list of vectors.'''
    
    reviews_vec = []

    for review in reviews:
       
        ############### ADD YOUR CODE HERE ###############
        vector = [0]*len(vocabulary)
        counts = Counter(review)
        for i, term in enumerate(vocabulary):
            vector[i] = counts[term]
        reviews_vec.append(vector)
        ############### ADD YOUR CODE HERE ###############

    return reviews_vec

neg_reviews_train_vec = transform_to_vector(neg_reviews_train)
pos_reviews_train_vec = transform_to_vector(pos_reviews_train)
```

### Task 3 
Aggregate the vectors of the positive and negative reviews of the training set into a single vector. This vector will be the representative vector of each of the two sentiments. Which aggregation function did you choose and why?

```python
def aggregate_vector_list(vlist, aggfunc):
    if aggfunc == 'max':
        return np.array(vlist).max(axis=0)
    elif aggfunc == 'min':
        return np.array(vlist).min(axis=0)
    elif aggfunc == 'mean':
        return np.array(vlist).mean(axis=0)
    else:
        return np.zeros(np.array(vlist).shape[1])

############### ADD YOUR CODE HERE ###############
neg_reviews_repr = []
pos_reviews_repr = []

neg_reviews_repr = aggregate_vector_list(neg_reviews_train_vec, 'mean')
pos_reviews_repr = aggregate_vector_list(pos_reviews_train_vec, 'mean')
############### ADD YOUR CODE HERE ###############
```

Answer:

The presence of certain words represent a positive (resp. negative) sentiment. Hence, choosing the 'min' function would imply losing this information. Furthermore, some people may prefer to use 'sad' and others might prefer to use 'upset', so choosing 'min' function would also result in losing this information. The problem with 'max' is that not all the sentences use all the *positive* (resp. *negative*) words. Also, the lack of negative (positive) words might also represent a positive (negative) sentiment, and we lose this information with 'max'.

I decided to go with 'mean'.

### Task 5
Consider the following Naive Bayes sentiment classifier:

\begin{equation*}
P(sentiment | review)   = \frac{P(review | sentiment) * P(sentiment)}{P(review)}
\end{equation*}

Assume that you can estimate the likelihood probability as follows:

\begin{equation*}
P(review | sentiment) = cosine\_similarity(review\_vec, sentiment\_representative\_vec)
\end{equation*}

1) How do you take into account the probabilities: $P(sentiment)$ and $P(review)$?

Answer:

Cosine similarity compares two vectors by considering the angle between both in an N-dimensional space. 

For $P(sentiment)$ , to account for a sentiment being more likely than the other, you calculate the proportion of reviews having a certain sentiment in your train set, and you assume for the test set the proportion of each sentiment will be the same as for the train set (that is, for correct train and test sets). 

For $P(review)$, you could calculate the probability of each review.

2) Implement the classifier and compute its performance in terms of false negatives, true negatives, false positives and true positives.

```python
fn = 0
tn = 0
fp = 0
tp = 0

def cosine_similarity(a,b):
    '''Computes the cosine similarity of $a$ and $b$.'''
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)
    
############### ADD YOUR CODE HERE ###############

for review_v in neg_reviews_test_vec:
    tag_idx = np.argmax([cosine_similarity(review_v, neg_reviews_repr),
        cosine_similarity(review_v, pos_reviews_repr)])
    tag = ['N', 'P'][tag_idx]
    if tag == 'N':
        tn += 1
    else:
        fp += 1
    

for review_v in pos_reviews_test_vec:
    tag_idx = np.argmax([cosine_similarity(review_v, neg_reviews_repr),
        cosine_similarity(review_v, pos_reviews_repr)])
    tag = ['N', 'P'][tag_idx]
    if tag == 'P':
        tp += 1
    else:
        fn += 1
        
print('tp',tp,', tn', tn, ', fn', fn, ', fp', fp)
```

### Task 6
